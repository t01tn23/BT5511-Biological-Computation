# Import, append and concatenate all data files into one panda dataframe

```python
import pandas as pd
import glob

dir_path = "C:/Users/Surface Book/Desktop/python/assessment analysis/data files"

def load_data(dir_path):
    files = glob.glob(dir_path + '/*.txt')

    all_data = []

    for f in files:
        txt_files = pd.read_csv(f, sep='\t')
        all_data.append(txt_files)
    
    return(all_data)
        
all_data = load_data(dir_path)   

concated_data = pd.concat(all_data) 
concated_data
```

```python
#check the data set type to ensure the output is correct
type(concated_data)
```

# Averaging each three replicates mRNA expression, protein expression and cell viability quantifications for wildtype and mutant cells

```python
#separate columns for mutant and wildtype as well as their quantification type (mRNA/protein/cell viability)
gene_name = concated_data.loc[:, ['Gene']]
WT_mRNA_columns = concated_data.loc[:, ['mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3']]
mut_mRNA_columns = concated_data.loc[:, ['mRNA.Expression.Mut.Rep1', 'mRNA.Expression.Mut.Rep2', 'mRNA.Expression.Mut.Rep3']]
WT_protein_columns = concated_data.loc[:, ['Protein.Expression.WT.Rep1', 'Protein.Expression.WT.Rep2', 'Protein.Expression.WT.Rep3']]
mut_protein_columns = concated_data.loc[:, ['Protein.Expression.Mut.Rep1', 'Protein.Expression.Mut.Rep2', 'Protein.Expression.Mut.Rep3']]
WT_cellvia_columns = concated_data.loc[:, ['CellViability.WT.Rep1', 'CellViability.WT.Rep2', 'CellViability.WT.Rep3']]	
mut_cellvia_columns = concated_data.loc[:, ['CellViability.Mut.Rep1', 'CellViability.Mut.Rep2', 'CellViability.Mut.Rep3']]

#calculate averages for each row
WT_mRNA_avg = WT_mRNA_columns.mean(axis=1)
mut_mRNA_avg = mut_mRNA_columns.mean(axis=1)
WT_protein_avg = WT_protein_columns.mean(axis=1)
mut_protein_avg = mut_protein_columns.mean(axis=1)
WT_cellvia_avg = WT_cellvia_columns.mean(axis=1)
mut_cellvia_avg = mut_cellvia_columns.mean(axis=1)

#concatenate columns back together and add a title to the column
all_averages = pd.concat([gene_name, WT_mRNA_avg, mut_mRNA_avg, WT_protein_avg, mut_protein_avg, WT_cellvia_avg, mut_cellvia_avg], axis=1)
all_averages.columns = ['Gene', 'WT average mRNA expression', 'MUT average mRNA expression', 'WT average protein expression', 'MUT average protein expression', 'WT average cell viability', 'MUT average cell viability']
all_averages
```

# Mutation class identification

```python
#count characters within wildtype and mutant gene sequences for each each
WT_seq_count = concated_data['WildType.Sequence'].str.len()
MUT_seq_count = concated_data['Mutant.Sequence'].str.len()

#define a function to apply a conditional statements that identifies mutation class
def mutant_type(WT, MUT):
    mutation_types = []
    for wt, mut in zip(WT, MUT):
        if wt > mut:
            mutation_types.append('deletion')
        elif wt < mut:
            mutation_types.append('insertion')
        elif wt == mut:
           mutation_types.append('substitution')
    return mutation_types

mutation_types = mutant_type(WT_seq_count, MUT_seq_count)
mutation_type_df = pd.DataFrame(mutation_types) #changed list into a dataframe, error code showed that it was not compatible to concatenate using pandas
mutation_type_df.columns = ['Mutation Class'] #added a column title 
mutation_type_df

#concat mutant type data frame to the all_averages data frame to make a summary table
summary = pd.concat([all_averages, mutation_type_df],axis=1)
summary
```

```python
#sorted mutations into their classes to identify genes of interest more easily
mutant_sorted_df = summary.sort_values(by='Mutation Class')
mutant_sorted_df
```

# Calculating the difference between wildtype and mutant mRNA expression, protein expression and cell viability

```python
#calculating the difference between WT and MUT mRNA expression, protein expression and cell viability.
gene = mutant_sorted_df['Gene']
mutant_class = mutant_sorted_df['Mutation Class']
difference_mRNA = mutant_sorted_df['WT average mRNA expression'] - mutant_sorted_df['MUT average mRNA expression'] 
difference_protein = mutant_sorted_df['WT average protein expression'] - mutant_sorted_df['MUT average protein expression']
difference_cellvia = mutant_sorted_df['WT average cell viability'] - mutant_sorted_df['MUT average cell viability']

difference_data = pd.concat([gene, mutant_class, difference_mRNA, difference_protein, difference_cellvia], axis=1)
difference_data.columns = ['Gene','Mutation Class', 'mRNA Expression Difference', 'Protein Expression Difference', 'Cell Viability Difference']
difference_data
```
